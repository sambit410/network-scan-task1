```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Wireshark CSV
df = pd.read_csv('/content/wireshark.csv')

df['Time'] = pd.to_datetime(df['Time'], unit='s', origin='unix')

packet_counts = df.resample('S', on='Time').size()


plt.figure(figsize=(10, 6))
packet_counts.plot()
plt.title('Packet Count Over Time')
plt.xlabel('Time')
plt.ylabel('Number of Packets')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot histogram of packet sizes
plt.figure(figsize=(10, 6))
df['Length'].hist(bins=50)
plt.title('Packet Size Distribution')
plt.xlabel('Packet Size (Bytes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Count packets by protocol
protocol_counts = df['Protocol'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
protocol_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Protocol Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
